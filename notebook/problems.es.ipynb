## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

## Explicacion del código

Utilizamos random.randint para generar 1000 tiradas de cada dado de numeros aleatorios del 1 al 6.

Guardamos en el diccionario 'tiradas' un array para cada clave con la suma de las tiradas de cada dado mayor que 7 (clave 'mayor7') y par (clave 'par').

Guardamos en el diccionario 'mayorypar7':

- lanzamientos>7 :  numero de ocurrencias que salio la suma de los dos dados mayor que 7.
- lanzamientos_par: numero de ocurrencias que salio la suma de los dos dados par.
- probabilidad>7: probabilidad de que salga mayor que 7 en la suma de los dos dados.
- probabilidad>par: probabilidad de que salga par en la suma de los dos dados.


```python
lanzamientos = 1000
dado1 = np.random.randint(1, 7, lanzamientos)
dado2 = np.random.randint(1, 7, lanzamientos)
suma_dados = dado1 + dado2

tiradas = {'mayor7':suma_dados[suma_dados > 7],
           'par':suma_dados[suma_dados %2 == 0]}

mayor7_par = {'lanzamientos>7': len(tiradas['mayor7']),
              'lanzamientos_par': len(tiradas['par']),
              'probabilidad>7': round((len(tiradas['mayor7'])) / lanzamientos, 2),
              'probabilidad_par': round((len(tiradas['par'])) / lanzamientos, 2)}

mayor7_par
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

## Explicacion del código

Utilizamos el metodo random.choice para generar un array de 5000 bolas de los tres colores intriduciondole la probabilidad de que salga cada color (cincomil_bolas)

Utilizamos un bucle para convertir el array de 5000 elementos en una lista de 5000 elementos repartidos en sublistas de 5 arrays (cincomil_bolas_sublist5)

Utilizamos otro bucle para recorrer cada sublista de 5 elementos y comparamos utilizando slicing:
- Apartado1: los 3 primeros elementos y los 2 ultimos de cada sublista de 5 elementos, si encuentra coincidencias las guardamos en un contador('coincidencias1')
- Apartado2: todos los elementos de cada , si encuentra coincidencias las guardamos en un contador('coincidencias2')

Tambien guardamos una lista de coincidencias para comprobar los resultados

Mostramos un diccionaro para cada resultado obtenido en cada apartado(Resultados_apartado1 y Resultados_apartado1)



```python
muestras = 5000
colors = ['white', 'red', 'green']
prob = [10/60, 20/60, 30/60]
cincomil_bolas = np.random.choice(colors, muestras, prob)
cincomil_bolas_sublist5 = []
lista_coincidencias1 = []
lista_coincidencias2 = []
coincidencias1 = 0
coincidencias2 = 0

for i in range (0, muestras, 5):
    cincomil_bolas_sublist5.append(list(cincomil_bolas[i:i+5]))

for sublist5 in cincomil_bolas_sublist5:
    
    if sublist5.count('white') == 3 and sublist5.count('red') == 2:
        coincidencias1 += 1
        lista_coincidencias1.append(sublist5)
    
    if sublist5.count('white') == 5 or sublist5.count('red') == 5 or sublist5.count('green') == 5:
        coincidencias2 += 1
        lista_coincidencias2.append(sublist5)


Resultados_apartado1 = {'Numero de coincidencias': coincidencias1,
                        'Lista de coincidencias': lista_coincidencias1,
                        'Probabilidad': coincidencias1 / muestras}


Resultados_apartado2 = {'Numero de coincidencias': coincidencias2,
                        'Lista de coincidencias': lista_coincidencias2,
                        'Probabilidad': coincidencias2 / muestras}


Resultados_apartado1
```
